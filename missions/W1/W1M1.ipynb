### gear, transmission(am) features

- gearì transmissionì ê²½ì°ì ì ì¡°í©ë³ë¡ ëëê³  ê° ì¡°í©ì ëª ê°ì ì°¨ëì´ ìëì§ íì¸ with Bar

```python
print(f"gearì ì¡´ì¬íë ê°ì ê°ì: {len(data['gear'].unique())}ê° {data['gear'].unique()}")
print(f"transmissionì ì¡´ì¬íë ê°ì ê°ì: {len(data['am'].unique())}ê° {data['am'].unique()}")
```

```python
# crosstab ì¬ì©

# crosstabì¼ë¡ gearì am ì¡°í©ë³ count ë§ë¤ê¸°
ct = pd.crosstab(index=data['gear'], columns=data['am'])

# stackí´ì (gear, am) íí ì¸ë±ì¤ë¡ ë§ë¤ê³  reset_index
df_ct = ct.stack().reset_index()
df_ct.columns = ['gear', 'am', 'count']

# ê²°ê³¼ê° ì¶ë ¥
print(df_ct)

# (gear, am) ííì Xì¶ ë¼ë²¨ë¡ ì¬ì©
df_ct['x_label'] = list(zip(df_ct['gear'], df_ct['am']))
x_pos = range(len(df_ct))

# ë§ëê·¸ëí ê·¸ë¦¬ê¸°
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x_pos, df_ct['count'])

# Xì¶ ë¼ë²¨ ì¤ì 
ax.set_xticks(x_pos)
ax.set_xticklabels(df_ct['x_label'], rotation=45)
ax.set_xlabel('(Gear, Transmission)')
ax.set_ylabel('# of Cars')
ax.set_title('# of Cars by Gear x Transmission')

plt.tight_layout()
plt.show()
```

```python
# groupby ì¬ì©

# ê°ë¥í gear, am ì¡°í© ìì±
all_combinations = pd.MultiIndex.from_product(
    [sorted(data['gear'].unique()), sorted(data['am'].unique())],
    names=['gear', 'am']
)

# groupby ì§ê³
grouped = data.groupby(['gear', 'am']).size().reindex(all_combinations, fill_value=0).reset_index(name='count')

# ê²°ê³¼ê° ì¶ë ¥
print(grouped)

# Xì¶ ë¼ë²¨ (gear, am íí)
x_labels = list(zip(grouped['gear'], grouped['am']))
x_pos = range(len(x_labels))

# ë§ëê·¸ëí
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x_pos, grouped['count'])

# Xì¶ ì¤ì 
ax.set_xticks(x_pos)
ax.set_xticklabels(x_labels, rotation=0)
ax.set_xlabel('(Gear, Transmission)')
ax.set_ylabel('# of Cars')
ax.set_title('# of Cars by Gear x Transmission')

plt.tight_layout()
plt.show()
```

### ëª¨ë  featuresì ëí íì¤í ê·¸ë¨

```python
# ìì¹í ì»¬ë¼ë§ ì í
numeric_cols = data.select_dtypes(include='number').columns
n_cols = len(numeric_cols)

# subplot í¬ê¸° ì¤ì  (í x ì´ ìë ê²°ì )
n_rows = int(n_cols / 3 + 0.5)  # 3ì´ë¡ ì ë ¬
fig, axes = plt.subplots(n_rows, 3, figsize=(15, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].hist(data[col], bins=10, color='skyblue', edgecolor='black')
    axes[i].set_title(f'{col}')

# 4. ë¨ì subplot ì¨ê¸°ê¸° (ë°ì´í° ì»¬ë¼ ìê° 3ì ë°°ìê° ìë ê²½ì° ëë¹)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.suptitle("Histograms of Variables", fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
